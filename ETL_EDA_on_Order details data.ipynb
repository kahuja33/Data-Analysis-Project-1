```python
import pandas as pd
df=pd.read_csv("orders.csv",na_values=['Not Available', 'unknown'])
df["Ship Mode"].unique()
```

```python
#df.rename(columns={"Order Id":"order_id"})
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df.head(5)
```

```python
df["discount"]=df["list_price"]*df["discount_percent"]*0.01

```

```python
df["sale_price"]=df["list_price"]-df["discount"]
df
```

```python
df["profit"]=df["sale_price"]-df["cost_price"]
```

```python
df.order_date=pd.to_datetime(df["order_date"],format="%Y-%m-%d")
```

```python
df.drop(columns=["list_price","cost_price","discount_percent"],inplace=True)
```

```python
#Top 10 highest selling Products

pd.read_sql_query("Select product_id,sum(sale_price*quantity) as sales from df_orders group by product_id order by sales desc limit 10",cnn)
```

```python
# Top 5 highest Selling Products in each region

pd.read_sql_query(
"With cte as (Select region,product_id,sum(Sale_price) as sales from df_orders group by region,product_id order by region,sales desc)"+
"Select * from (Select *,rank() over (partition by region order by sales desc) as rnk from cte) as temp where rnk<=5",cnn)
```

```python
# Find MoM growth Comparison for 2022 and 2023 sales

pd.read_sql_query("With cte as (Select month(order_date) as month,sum(sale_price) as sales from df_orders where year(order_date)=2022 group by month(order_date))"+
                  ",cte2 as (Select month(order_date) as month,sum(sale_price) as sales from df_orders where year(order_date)=2023 group by month(order_date))"+
                  "select a.month,a.sales,b.sales from cte a inner join cte2 b on a.month=b.month",cnn)
```

```python
# For each category which month has the highest sales

pd.read_sql_query("WITH cte AS (SELECT category,CONCAT(MONTH(order_Date), '-', YEAR(order_date)) AS date,SUM(sale_price) AS sales FROM df_orders GROUP BY category, CONCAT(MONTH(order_Date), '-', YEAR(order_date)))SELECT * FROM (SELECT *, RANK() OVER(PARTITION BY category ORDER BY sales DESC) AS rnk FROM cte) AS temp WHERE rnk = 1",cnn)
```

```python
# Which cateogry has highest growth by profit in 2023 compared 2022

pd.read_sql_query("With cte as (Select sub_category,sum(sale_price) as sales_2023 from df_orders where year(order_date)=2023 group by sub_category),cte2 as (Select sub_category,sum(sale_price) as sales_2022 from df_orders where year(order_date)=2022 group by sub_category)"+
                  "Select a.sub_category,a.sales_2023,b.sales_2022,(a.sales_2023-b.sales_2022)*100/sales_2022 as sales_diff_percent from cte a inner join cte2 b on a.sub_category=b.sub_category order by sales_diff_percent desc limit 1",cnn)
```
